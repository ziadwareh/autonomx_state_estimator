```python
import numpy as np
import scipy.fft
import matplotlib.pyplot as plt
```

```python
# sampling frequency
fs = 200

T = 1/fs

t = np.linspace(0,1,fs)
print(len(t))

signal = np.sin(2 * 2*np.pi * t) + 0.2*np.sin(50 * 2*np.pi * t)

fourier_transform = scipy.fft.fft(signal)
plt.plot(t,signal)
plt.grid()
plt.show()
```

```python
N = len(fourier_transform)
# f_axis = np.linspace(-fs/2,fs/2,N)
f_axis = np.linspace(-fs/2, fs/2, N)
magnitude_spectrum = np.abs(scipy.fft.fftshift(fourier_transform))
plt.plot(f_axis,magnitude_spectrum)
plt.grid()
```
